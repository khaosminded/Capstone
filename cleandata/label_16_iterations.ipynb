### import data

```r
NAMES <- read.table("./named/y_cruncher_dedicated_2ndrun.csv", nrow = 1, stringsAsFactors = FALSE, sep = ",")
DATA <- read.table("./named/y_cruncher_dedicated_2ndrun.csv", skip = 1, stringsAsFactors = FALSE, sep = ",")
names(DATA) <- NAMES 
data=DATA[,c('instanceID','testStartTime','computationTime','benchmarkTime','wallTime')]
```

### sort instanceID by overall comutationTime
### cause we know that tasks on every VM are 10 runs apart

```r
sum=aggregate(x=data["computationTime"],
          by=list(instanceID=DATA$instanceID),
          FUN=sum
         )
sorted=sum[order(sum$computationTime),] 
sortedInstanceID=sorted$instanceID
sortedInstanceID
sorted
```

### label 1st 10 runs on each VM as 'set 1'
### then  2nd 10 runs as 'set 2' ... 16th 10 runs as 'set 10'
### label the VM that only has 10 runs as 'VM 1' ... 
### which has 160runs is the 'VM 16'

```r
runs=10
labeled=data[0:0,]
InstanceId=sortedInstanceID
n=length(InstanceId)
for(j in 1:n){
    for (i in InstanceId){

        forthisInstance=data[data$instanceID==i,]
        x=(j-1)*runs+1
        y=(j*runs)
        forthisInstance=data.frame(forthisInstance[x:y,],'set'=j,'vmID'=which(InstanceId==i)+j-1)
        
        labeled<-rbind(labeled,forthisInstance) 
    }
    InstanceId=InstanceId[-1]
}

labeled
```

# output labeled data

```r
write.csv(labeled[,1:7],"y_cruncher_dedicated_2ndrun_labeled.csv")
```
